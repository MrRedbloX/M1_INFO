```python
import matplotlib.pyplot as plt
import random
import time
from itertools import permutations 
from scipy.special import comb

def main(n):
    M = 5*n;
    util = [random.randint(1,10) for i in range(0,n)]
    poids = [random.randint(1,10) for i in range(0,n)]
    
    print("Méthode brute force sans répétition : ")
    print("Gain :",bruteForce(M,poids.copy(), util.copy(),len(poids),True))
    print("Méthode brute force avec répétition : ")
    print("Gain :",bruteForce(M,poids.copy(), util.copy(),len(poids),False))
    
    print("Méthode gloutonne sans répétition : ")
    print("Gain :", gloutonne(M,poids.copy(), util.copy(),True))
    print("Méthode gloutonne avec répétition : ")
    print("Gain :", gloutonne(M,poids.copy(), util.copy(),False))

    print("Méthode dynamique : ")
    print("Gain :", dynamique(M,poids.copy(), util.copy(),len(poids)))
    
def bruteForce(W , wt , val, n, unique):
    if(unique): 
        if n == 0 or W == 0 : 
            return 0
        if (wt[n-1] > W): 
            return bruteForce(W , wt , val , n-1, True) 
        else: 
            return max(val[n-1] + bruteForce(W-wt[n-1] , wt , val , n-1,True), 
                       bruteForce(W , wt , val , n-1,True)) 
    else:
        dp = [0 for i in range(W + 1)] 
        ans = 0
        for i in range(W + 1): 
            for j in range(n): 
                if (wt[j] <= i): 
                    dp[i] = max(dp[i], dp[i - wt[j]] + val[j]) 
        return dp[W] 

def gloutonne(M, vP, vU, unique):
    vRapport = [vU[i]/vP[i] for i in range(0,len(vU))]
    gain = 0
    remplissage=0
    while(len(vRapport)>0 and remplissage<M):
        maxIndex = np.argmax(vRapport);
        if(vP[maxIndex]+remplissage<=M):
            remplissage = remplissage+vP[maxIndex]
            gain = gain+vU[maxIndex]
            if(unique):
                vP.pop(maxIndex)
                vU.pop(maxIndex)
                vRapport.pop(maxIndex)        
        else:
            vP.pop(maxIndex)
            vU.pop(maxIndex)
            vRapport.pop(maxIndex)
    return gain;

def dynamique(W, wt, val, n): 
    K = [[0 for x in range(W+1)] for x in range(n+1)] 
    for i in range(n+1): 
        for w in range(W+1): 
            if i==0 or w==0: 
                K[i][w] = 0
            elif wt[i-1] <= w: 
                K[i][w] = max(val[i-1] + K[i-1][w-wt[i-1]],  K[i-1][w]) 
            else: 
                K[i][w] = K[i-1][w] 
    return K[n][W]

main(20)

def mainWithTime(n):
    M = 5*n;
    vUtil = [random.randint(1,10) for i in range(0,n)]
    vPoids = [random.randint(1,10) for i in range(0,n)]
    resBruteForce=[]
    resBruteForce2=[]
    resGlouton=[]
    resGlouton2=[]
    resDynamic=[]
    for j in range(1,n):
        moyBruteForce=0
        moyBruteForce2=0
        moyGlouton=0
        moyGlouton2=0
        moyDynamic=0
        for k in range(1,int(n/2+1)):
            t1 = time.process_time()
            bruteForce(5*j, vPoids[:j] , vUtil[:j], j, True)
            t2 = time.process_time()
            moyBruteForce=moyBruteForce+(t2-t1)
            
            t1 = time.process_time()
            bruteForce(5*j, vPoids[:j] , vUtil[:j], j, False)
            t2 = time.process_time()
            moyBruteForce2=moyBruteForce2+(t2-t1)
            
            
            t1 = time.process_time()
            gloutonne(5*j, vPoids[:j] , vUtil[:j], True)
            t2 = time.process_time()
            moyGlouton=moyGlouton+(t2-t1)
            
            t1 = time.process_time()
            gloutonne(5*j, vPoids[:j] , vUtil[:j], False)
            t2 = time.process_time()
            moyGlouton2=moyGlouton2+(t2-t1)
            
            t1 = time.process_time()
            dynamique(5*j, vPoids[:j] , vUtil[:j], j)
            t2 = time.process_time()
            moyDynamic=moyDynamic+(t2-t1)
            
            
            
        moyBruteForce=moyBruteForce/10
        moyBruteForce2=moyBruteForce2/10
        moyGlouton=moyGlouton/10
        moyGlouton2=moyGlouton2/10
        moyDynamic=moyDynamic/10
        
        resBruteForce.append(moyBruteForce)
        resBruteForce2.append(moyBruteForce2)
        resGlouton.append(moyGlouton)
        resGlouton2.append(moyGlouton2)
        resDynamic.append(moyDynamic)
        
    plt.figure(figsize=(6, 6))
    plt.plot(np.arange(1,n), resBruteForce,label="brute")
    plt.plot(np.arange(1,n), resBruteForce2,label="brute avec répétition")
    plt.plot(np.arange(1,n), resGlouton,label="glouton")
    plt.plot(np.arange(1,n), resGlouton2,label="glouton avec répétition")
    plt.plot(np.arange(1,n), resDynamic,label="dynamique")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()
    
    
    resGlouton=[]
    resGlouton2=[]
    resDynamic=[]
    for j in range(1,n):
        moyGlouton=0
        moyGlouton2=0
        moyDynamic=0
        for k in range(1,int(n/2+1)):
            t1 = time.process_time()
            gloutonne(5*j, vPoids[:j] , vUtil[:j], True)
            t2 = time.process_time()
            moyGlouton=moyGlouton+(t2-t1)
            
            t1 = time.process_time()
            gloutonne(5*j, vPoids[:j] , vUtil[:j], False)
            t2 = time.process_time()
            moyGlouton2=moyGlouton2+(t2-t1)
            
            t1 = time.process_time()
            dynamique(5*j, vPoids[:j] , vUtil[:j], j)
            t2 = time.process_time()
            moyDynamic=moyDynamic+(t2-t1)

        moyGlouton=moyGlouton/10
        moyGlouton2=moyGlouton2/10
        moyDynamic=moyDynamic/10
        resGlouton.append(moyGlouton)
        resGlouton2.append(moyGlouton2)
        resDynamic.append(moyDynamic)
        
    plt.figure(figsize=(6, 6))
    plt.plot(np.arange(1,n), resGlouton,label="glouton")
    plt.plot(np.arange(1,n), resGlouton2,label="glouton avec répétition")
    plt.plot(np.arange(1,n), resDynamic,label="dynamique")
    plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.show()
    
mainWithTime(20)
```
