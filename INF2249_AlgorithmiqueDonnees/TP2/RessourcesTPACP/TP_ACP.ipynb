### Donn√©es centr√©es et donn√©es centr√©es-r√©duites

* Visualiser les variables ``Intensite odeur``, ``Glucose`` et ``Acide citrique`` pour les 6 √©chantillons du jeux de donn√©es

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D  
```

```python
def viz3D(data):
    fig = plt.figure(figsize=(8,6))
    ax = Axes3D(fig)

    dx = data['Intensite odeur']
    dy = data['Glucose']
    dz = data['Acide citrique']

    ax.scatter(dx, dy, dz, s=50, edgecolors='w')

    for i, p in enumerate(data.index):
        ax.text(dx[i]+ 0.01, dy[i]+0.01, dz[i]+0.01, p, fontsize=12)

    ax.set_xlabel('Intensite odeur')
    ax.set_ylabel('Glucose')
    ax.set_zlabel('Acide citrique')

    plt.show()
viz3D(orange)
```

* Centrer l'ensemble des donn√©es (soustraire la moyenne) (cr√©er une variable ``Y``)
* Revisualiser les donn√©es

```python
Y = orange - orange.mean()
Y
viz3D(Y)
```

* Centrer et r√©duire l'ensemble des donn√©es (soustraire la moyenne et diviser par l'√©cart-type) (cr√©er une variable ``Z``)
* Revisualiser les donn√©es

```python
Z = Y / orange.std(ddof=0)
Z
viz3D(Z)
```

### ACP sur les donn√©es centr√©es (``Y``)

```python
# Importer le paquetage PCA de la biblioth√®que decomposition de Scikit-Learn
from sklearn.decomposition import PCA
```

* Lire la documentation de PCA [en ligne](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html)


Essayer d'effectuer l'ensemble des actions suivantes
1. Visualiser la valeurs des valeurs propres
2. Visualiser la valeur cumul√©e des valeurs propres
3. Afficher la valeur des vecteurs propres
4. Visualiser la projection des donn√©es sur les deux premi√®res dimensiosn.

```python
pca = PCA()
comp = pca.fit_transform(data)
df = pd.DataFrame(data = comp, columns = data.columns, index=data.index)
df

```

```python
import seaborn as sns
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform

pairwise = pd.DataFrame(
    squareform(pdist(df)),
    columns = df.index,
    index = df.index
)
sns.heatmap(
    pairwise,
    cmap='OrRd',
    linewidth=1,
    square=True
)
```

## ACP norm√©e
Recommencer les √©tapes pr√©c√©dentes sur les donn√©es centr√©es r√©duites (``Z``)

```python
Y = df - df.mean()
Z = Y / df.std(ddof=0)

pairwiseZ = pd.DataFrame(
    squareform(pdist(Z)),
    columns = Z.index,
    index = Z.index
)
sns.heatmap(
    pairwiseZ,
    cmap='OrRd',
    linewidth=1,
    square=True
)
```

Est ce que les donn√©es projet√©es sont similaires pour l'ACP et l'ACP norm√©e ?

Pour l'ACP de base, les donn√©es d'une certaine cat√©gorie contiennent des valeurs qui sont tr√®s grandes par rapport aux autres valeurs de cette cat√©gorie. On observe alors de gros √©carts dans le heatmap.
Lorsqu'on centre l'ACP, les donn√©es sont plus proche et on arrive √† observer des √©carts qui ont du sens.
